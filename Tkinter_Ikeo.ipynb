```python
import tkinter as tk
from tkinter import *
import mysql.connector
from PIL import Image, ImageTk

config = {
      'user': 'root',
      'password': 'root',
      'host': 'localhost',
      'port': '8081',
      'database': 'ikeof',
      'raise_on_warnings': True,
    }

#Création du Cursor
link = mysql.connector.connect(**config)
cursor = link.cursor()

#Création de la liste de produit dans le menu
def menu_deroulant():
    new_list = []
    new_new_list = []
    cursor.execute("SELECT id_p, nom_p FROM produits ")
    rows_produits = cursor.fetchall()
    
    for produit in rows_produits:
        new_list.append(produit[0])
        new_new_list.append(produit[1])
    return new_list, new_new_list

#Création de l'interface graphique        
app = tk.Tk()
app.geometry('700x700')
app.config(background='#CE0036')

#Images
logo = Image.open('ikeo_titre2.png')
logo = ImageTk.PhotoImage(logo)
logo_label = tk.Label(bd = 0, image=logo)
logo_label.image= logo
logo_label.pack()

variable = tk.StringVar(app)
variable.set("")

#Incorporation de la liste dans le menu déroulant
opt = tk.OptionMenu(app, variable, *menu_deroulant()[1])
opt.config(width=13, font=('Consolas', 15), bg="#CE0036", fg="yellow", activebackground="#CE0036", activeforeground="yellow", highlightthickness=0)
opt.pack(side="top")

#Texte contenant la liste des usines
labelTest = tk.Label(text="", font=('Consolas', 15))
labelTest.pack(side="top")

#Création de la liste usine en fonction des produits
def usines():
    new_list = []
    cursor.execute(f"SELECT `nom_u`,`adresse_u`,`ville_u` FROM usines JOIN produits_usines ON usines.id_u = produits_usines.usine_id JOIN produits ON produits_usines.produit_id=produits.id_p WHERE nom_p = '{variable.get()}'")
    rows_usines = cursor.fetchall()
    
    for items in rows_usines: 
        new_list.append(items)
    return new_list  

#Affichage texte contenant la liste des usines
def callback(*args):
    labelTest.configure(text=f"{usines()}", font=('Consolas', 13), fg='yellow', bg="#CE0036", padx=15, pady=15)
    
variable.trace("w", callback)



#Affichage de l'interface graphique
app.mainloop()
```
